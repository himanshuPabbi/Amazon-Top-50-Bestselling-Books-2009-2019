```python
q1=df['User Rating'].quantile(0.25)
```

```python
q3=df['User Rating'].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
lower=q1-1.5*IQR
```

```python
upper=q3+1.5*IQR
```

```python
df=df[(df['User Rating']>lower) & (df['User Rating']<upper)]
```
